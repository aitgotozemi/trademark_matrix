# 00 DB 接続確認

データベースへの接続確認と、区分数ごとの出願件数を集計する。

```python
from utils.db import get_conn

conn = get_conn()
cur = conn.cursor()
```

```python
cur.execute("""
    SELECT "number-of-goods-service-class", COUNT(*) AS cnt
    FROM trademark_db
    GROUP BY "number-of-goods-service-class"
    ORDER BY "number-of-goods-service-class"
""")
for row in cur.fetchall():
    print(f'区分数: {row[0]}  件数: {row[1]}')
```

```python
conn.close()
```
