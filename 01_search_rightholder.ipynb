# 商標権利者 キーワード検索

商標権者氏名（`right-holder-group-name`）および商標権利者識別番号（`right-holder-group-identification-number`）によるキーワード検索ノートブック。

```python
import pandas as pd
from utils.db import get_conn, DB_CONFIG

print('DB設定を読み込みました:', DB_CONFIG['host'], DB_CONFIG['dbname'])
```

---

## 1. 商標権者氏名で検索

`right-holder-group-name` を部分一致（ILIKE）で検索し、権利者ごとの出願件数サマリを表示します。

```python
# ===== 検索キーワードをここに入力 =====
name_keyword = 'ベネッセ'   # 例: 'トヨタ', 'ソニー', 'Toyota'
# =====================================

with get_conn() as conn:
    df = pd.read_sql("""
        SELECT
            \"right-holder-group-identification-number\",
            \"right-holder-group-name\",
            \"application-number\",
            \"number-of-goods-service-class\"
        FROM trademark_db
        WHERE \"right-holder-group-name\" ILIKE %(keyword)s
    """, conn, params={'keyword': f'%{name_keyword}%'})

summary_name = (
    df
    .groupby(['right-holder-group-identification-number', 'right-holder-group-name'])
    .agg(
        出願件数=('application-number', 'count'),
        平均区分数=('number-of-goods-service-class', 'mean'),
        最大区分数=('number-of-goods-service-class', 'max'),
    )
    .reset_index()
    .sort_values('出願件数', ascending=False)
)
summary_name['平均区分数'] = summary_name['平均区分数'].round(2)

print(f'検索キーワード: 「{name_keyword}」  ヒット権利者数: {len(summary_name):,} 件')
display(summary_name)
```

---

## 2. 商標権利者識別番号で検索

`right-holder-group-identification-number` を完全一致で検索し、権利者ごとの出願件数サマリを表示します。

```python
# ===== 識別番号をここに入力 =====
identification_number = '000003207'   # 例: '000003207'
# ================================

with get_conn() as conn:
    df2 = pd.read_sql("""
        SELECT
            \"right-holder-group-identification-number\",
            \"right-holder-group-name\",
            \"application-number\",
            \"number-of-goods-service-class\"
        FROM trademark_db
        WHERE \"right-holder-group-identification-number\" = %(id)s
    """, conn, params={'id': identification_number})

summary_id = (
    df2
    .groupby(['right-holder-group-identification-number', 'right-holder-group-name'])
    .agg(
        出願件数=('application-number', 'count'),
        平均区分数=('number-of-goods-service-class', 'mean'),
        最大区分数=('number-of-goods-service-class', 'max'),
    )
    .reset_index()
    .sort_values('出願件数', ascending=False)
)
summary_id['平均区分数'] = summary_id['平均区分数'].round(2)

print(f'識別番号: 「{identification_number}」  ヒット権利者数: {len(summary_id):,} 件')
display(summary_id)
```
